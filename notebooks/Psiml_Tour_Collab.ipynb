Google Colab is an online platform for running Python and Jupyter notebooks without any installation. It provides a preconfigured environment with free GPU/TPU resources for machine learning and data analysis.

You can run simple python code

```python
print("Hello from Google Colab!")
```

It comes preinstalled with most commonly used libraries.

```python
import numpy as np

a = np.array([1, 2, 3])
print("Array:", a)
print("Mean:", a.mean())
```

Even libraties for Machine learning

```python
import torch

print("GPU available:", torch.cuda.is_available())
print("GPU name:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "No GPU")
```
